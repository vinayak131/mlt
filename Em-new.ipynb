```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import completeness_score
```

```python
iris  = load_iris()
values = iris.data
target = iris.target
values
target
```

```python
colormap = np.array(['red','lime','black'])
plt.title('Actual Values')
plt.scatter(values[:,2],values[:,3],c=colormap[target])
```

```python
# K-Means Plot
model=KMeans(n_clusters=3).fit(values)
plt.title('KMeans')
plt.scatter(values[:,2],values[:,3],c=colormap[model.labels_])
print('The completeness score of K-Means: ',completeness_score(target, model.labels_))
```

```python
#EM plot
gmm=GaussianMixture(n_components=3).fit(values)
target_pred=gmm.predict(values)
plt.title('EM')
plt.scatter(values[:,2],values[:,3],c=colormap[target_pred])
print('The completeness score of EM: ',completeness_score(target, target_pred))
```
