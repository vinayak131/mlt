```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import completeness_score
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
x = pd.DataFrame(iris.data)
x.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']
y = pd.DataFrame(iris.target)
y.columns = ['Targets']
```

```python
colormap = np.array(['red','lime','black'])
#Actual Plot
plt.scatter(x.Petal_Length,x.Petal_Width,c = colormap[y.Targets])
plt.title('Actual PLot')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
 #KMeans
model = KMeans(n_clusters=3)
model.fit(x)
plt.scatter(x.Petal_Length,x.Petal_Width,c=colormap[model.labels_])
plt.title('KMeans Plot')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
print('Completeness Score Kmeans : ',completeness_score(y.Targets,model.labels_))
```

```python
#Em
gm = GaussianMixture(n_components=3)
gm.fit(x)
y_pred = gm.predict(x)
plt.scatter(x.Petal_Length,x.Petal_Width,c=colormap[y_pred])
plt.title('EM PLot')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
print('Completeness Score EM : ',completeness_score(y.Targets,y_pred))
```
